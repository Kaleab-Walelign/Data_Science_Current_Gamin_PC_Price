```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
csv_file = open('Qefira_GamingPcs.csv','w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['pc_brand','condition','discription','price'])

for x in range(1,6):
    source = requests.get(f'https://www.qefira.com/classifieds?q=Gaming+laptop&page={x}').text
    soup = BeautifulSoup(source,'lxml')

    for listing_card in soup.find_all('div', class_='listing-card'):
        #title = listing_card['data-t-listing_title']
        
        try:
            brand = listing_card.find('span', class_='listing-card__header__tags__item--brand').text.strip()
        except Exception as e:
            brand = None
        
        try:
            price = listing_card.find('span', class_='listing-card__price__value').text.strip()
        except Exception as e:
            price = None
        try:
            condition = listing_card.find('span', class_='listing-card__header__tags__item--condition').text.strip()
        except Exception as e:
            condition = None
        
        try:
            url = listing_card.find('a', class_='listing-card__inner')['href']
        except Exception as e:
            url = None

        #print(f'Title: {title}')
        print(brand)
        print(price)
        print(condition)
        print(url)


        # Perform web scraping on the URL
        
        try:
            response = requests.get(url).text
            url_soup = BeautifulSoup(response, 'lxml')

            to_be_describe = url_soup.find('div', class_='listing-item__info')
            #print(to_be_describe.prettify())
            description = to_be_describe.find('div', class_='listing-item__description').find_all('p')
        
        except Exception as e:
            description = None
            
        print(description)
        print()
        csv_writer.writerow([brand,condition,description,price])
        

csv_file.close()
```
